```python
import pandas as pd
```

```python
categorie = pd.read_csv('categorie.csv')
client = pd.read_csv('client.csv')
commandes = pd.read_csv('commandes.csv')
produit = pd.read_csv('produit.csv')
sous_categorie = pd.read_csv('sous_categorie.csv')
```

```python
sous_categorie = pd.merge(sous_categorie, categorie, on='ID Categorie')
```

```python
commandes['Prix Unitaire'] = commandes['Prix Unitaire'].str.replace(',', '.').astype(float)
```

```python
commandes['prix totale'] = commandes['Quantite'] * commandes['Prix Unitaire']
```

```python
commandes
```

```python
produit['Cout Standard'] = produit['Cout Standard'].str.replace(',', '.').astype(float)
produit['Prix Standard'] = produit['Prix Standard'].str.replace(',', '.').astype(float)
```

```python
produit['Profit Total'] = produit['Prix Standard'] - produit['Cout Standard']
```

```python
commandes = pd.merge(commandes, produit, on='ID Produit')
```

```python
commandes = pd.merge(commandes, client, on='ID Client')
```

```python
commandes = pd.merge(commandes, sous_categorie, on='ID Sous Categorie')
```

```python
commandes['Date Commande'] = pd.to_datetime(commandes['Date Commande'])

commandes['Date Expedition'] = pd.to_datetime(commandes['Date Expedition'])
```

```python
commandes['Delai'] = commandes['Date Expedition'] - commandes['Date Commande']
```

```python
# afficher juste le nombre de jours pour le delai
commandes['Delai'] = commandes['Delai'].dt.days
```

```python
commandes.columns
```

```python
# replace les colunms dans un ordre logique 
commandes = commandes[['ID Commande', 'ID Client', 'Nom', 'Prenom', 'ID Produit', 'Nom Produit', 'ID Categorie', 'Categorie', 'ID Sous Categorie', 'Sous Categorie', 'Quantite', 'Prix Unitaire', 'prix totale', 'Cout Standard', 'Prix Standard', 'Profit Total', 'Date Commande', 'Date Expedition', 'Delai']]
```

```python
commandes.rename(columns={'prix totale': 'Prix Totale'}, inplace=True)
```
