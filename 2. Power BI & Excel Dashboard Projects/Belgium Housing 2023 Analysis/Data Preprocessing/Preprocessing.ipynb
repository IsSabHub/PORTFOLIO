```python
import pandas as pd
```

```python
df = pd.read_csv('combined_immo_be.csv')
```

```python
df['Price'] = df['Price'].str.replace('[\xa0 ]', '')
df['Price'] = df['Price'].astype(int)
```

```python
df.sort_values('Price', inplace=True, ascending=True)
```

```python
df['Surface'] = df['Surface'].str.replace('m²', '')
df['Surface'] = df['Surface'].astype(int)
```

```python
df['Price per m²'] = df['Price'] / df['Surface']
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
zip_code = pd.read_csv('code-postaux-belge.csv', sep=';')
zip_code
```

```python
postal_code = pd.read_csv('postal_codes.csv')
postal_code
```

```python
## merge postal_code and zip_code on 'Code' column
postal_code = pd.merge(zip_code,postal_code, on='Code', how='right')
```

```python
postal_code
```

```python
pp = pd.read_csv('postal-codes-belgium.csv', sep=';')
pp.drop(columns=pp.columns.difference(['Post code', 'Région name (French)']), inplace=True)
pp.rename(columns={'Post code':'Code', 'Région name (French)':'Region'}, inplace=True)
postal_code = pd.merge(postal_code, pp, left_on='Code', right_on='Code', how='left')
zip_code = pd.merge(zip_code, pp, left_on='Code', right_on='Code', how='left')
```

```python
postal_code
```

```python
df = pd.merge(df, postal_code, left_on='Location', right_on='Localite_y', how='left')
```

```python
df = pd.merge(df, zip_code, left_on='Location', right_on='Localite', how='left')
```

```python

# Fusionner les colonnes Localite_x et Localite_y
df["Code_x"].fillna(df["Code_y"], inplace=True)
df["Longitude_x"].fillna(df["Longitude_y"], inplace=True)
df["Latitude_x"].fillna(df["Latitude_y"], inplace=True)
df["Coordonnees_x"].fillna(df["Coordonnees_y"], inplace=True)
df["Region_x"].fillna(df["Region_y"], inplace=True)
```

```python
# Supprimer la colonne Localite_y
df.drop("Localite_y", axis=1, inplace=True)
df.drop("Localite_x", axis=1, inplace=True)
df.drop("Localite", axis=1, inplace=True)
df.drop("Code_y", axis=1, inplace=True)
df.drop("Longitude_y", axis=1, inplace=True)
df.drop("Latitude_y", axis=1, inplace=True)
df.drop("Coordonnees_y", axis=1, inplace=True)
df.drop("Region_y", axis=1, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
new_column_names = {'Title':'Type', 'Price':'Price', 'Surface':'Surface', 'Bedrooms':'Bedrooms', 'Bathrooms':'Bathrooms', 'Location':'City', 'Code_x':'Zip_code', 'Longitude_x':'Longitude', 'Latitude_x':'Latitude', 'Coordonnees_x':'Geographic_coordinates', 'Price per m²':'Price_per_m²', 'Region_x':'Region'}
```

```python
## rename columns
df.rename(columns=new_column_names, inplace=True)
```

```python
df = df.reindex(columns=['Type', 'Bedrooms', 'Bathrooms', 'Surface','Price', 'Price_per_m²', 'City', 'Zip_code','Region', 'Longitude', 'Latitude', 'Geographic_coordinates'])
```

```python
df['Zip_code'] = df['Zip_code'].astype(int)
```

```python
df['Price_per_m²'] = df['Price_per_m²'].round(2)
```

```python
df
```

```python
df.to_csv('Belgium_housing_dataset.csv', index=False)
```
